```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def plot(sort):
    plt.figure()
    fig, ax = plt.subplots(figsize=(20, 10))
    df = pd.read_csv(sort+".csv")
    plt.plot(df["Input Size"],df["Time taken(s)"])
    plt.xlabel("Input Size")
    plt.ylabel("Time taken(s)")
    plt.title(sort+" - Input Size vs Time taken")
    plt.savefig(sort+"_time")
    plt.figure()
    fig, ax = plt.subplots(figsize=(20, 10))
    plt.plot(df["Input Size"],df["#Comparisions"])
    plt.xlabel("Input Size")
    plt.ylabel("#Comparisions")
    plt.title(sort+" - Input Size vs #Comparisions")
    plt.savefig(sort+"_comparisions")
```

```python
plot("bubblesort")
```

```python
plot("selectionsort")
```

```python
plot("mergesort")
```

```python
plot("quicksort")
```

```python
def plot_overall():
    plt.figure()
    df1 = pd.read_csv("bubblesort.csv")
    df2 = pd.read_csv("selectionsort.csv")
    df3 = pd.read_csv("mergesort.csv")
    df4 = pd.read_csv("quicksort.csv")
    fig, ax = plt.subplots(figsize=(20, 10))
    plt.plot(df1["Input Size"],df1["Time taken(s)"])
    plt.plot(df2["Input Size"],df2["Time taken(s)"])
    plt.plot(df3["Input Size"],df3["Time taken(s)"])
    plt.plot(df4["Input Size"],df4["Time taken(s)"])
    plt.xlabel("Input Size")
    plt.ylabel("Time taken(s)")
    plt.title("Overall - Input Size vs Time taken")
    plt.legend(["bubble","selection","merge","quick"])
    plt.savefig("overall_time")
    plt.figure()
    fig, ax = plt.subplots(figsize=(20, 10))
    plt.plot(df1["Input Size"],df1["#Comparisions"])
    plt.plot(df2["Input Size"],df2["#Comparisions"])
    plt.plot(df3["Input Size"],df3["#Comparisions"])
    plt.plot(df4["Input Size"],df4["#Comparisions"])
    plt.legend(["bubble","selection","merge","quick"])
    plt.xlabel("Input Size")
    plt.ylabel("#Comparisions")
    plt.title("Overall - Input Size vs #Comparisions")
    plt.savefig("overall_comparisions")
```

```python
plot_overall()
```
